```python
def remove_normvector(set_acc, norm_uuid):
    raw_set = ff_utils.get_metadata(set_acc, key=my_auth, frame='raw')
    raw_pf = raw_set['processed_files']
    if norm_uuid in raw_pf:
        raw_pf.remove(norm_uuid)
        try:
            ff_utils.patch_metadata({'processed_files': raw_pf}, set_acc, key=my_auth)
            print('removed')
        except:
            print(set_acc, 'can not be patched')
         


print len(run_sets)

for a_set in run_sets:
    print(a_set['accession'])
    proc_f = a_set.get('processed_files')
    o_proc_f = a_set.get('other_processed_files')
    if proc_f:
        for a_fil in proc_f:
            if 'normvector_juicerformat' in a_fil['file_type_detailed']:
                print('found one in pf')
                remove_normvector(a_set['accession'], a_fil['uuid'])

                
    
    # see if anything on other processed files
    if o_proc_f:
        for f_set in o_proc_f:
            f_set_files = f_set['files']
            for a_file in f_set_files:
                if 'normvector_juicerformat' in a_file['file_type_detailed']:
                    print('found one in opf')
    print
    
```
